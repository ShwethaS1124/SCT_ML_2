```python
import numpy as np 
import pandas as pd 
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler as Ss
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Mall_Customers.csv')
```

```python
features=['Age','Annual Income (k$)','Spending Score (1-100)']
```

```python
sc=Ss()
```

```python
sc_data=sc.fit_transform(data[features])
```

```python
alg=KMeans(n_clusters=2)
alg.fit(sc_data)
```

```python
data['cluster']=alg.labels_
print(data['cluster'])
```

```python
plt.scatter(x=data['Age'],y=data['Annual Income (k$)'],c=data['cluster'])
```

```python
plt.scatter(x=data['Age'],y=data['Spending Score (1-100)'],c=data['cluster'])
```
